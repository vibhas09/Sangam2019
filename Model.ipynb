```python
train['date_time'] = pd.to_datetime(train['date_time'])
train.index = train['date_time']
```

```python
train.index[0].hour
```

```python

train['dayofweek'] = train.index.dayofweek
train['hour'] = train.index.hour
train['dayofmonth'] = train.index.day
train['dayofyear'] = train.index.dayofyear
train['month'] = train.index.month
train['year'] = train.index.year





```

```python
cc = train.groupby(['hour'])['traffic_volume'].mean().round(0)
```

```python
train['hour_mean'] = train['hour'].map(cc)
```

```python
dd = train.groupby(['is_holiday'])['traffic_volume'].mean().round(0)
dd
```

```python
train['is_holiday_mean'] = train['is_holiday'].map(dd)
```

```python
ee = train.groupby(['dayofweek'])['traffic_volume'].mean().round(0)
train['day_of_week_mean'] = train['dayofweek'].map(ee)
ff = train.groupby(['dayofmonth'])['traffic_volume'].mean().round(0)
train['day_of_month_mean'] = train['dayofmonth'].map(ff)
def isweekend(x):
    if (x == 5) or (x == 6):
        return 1
    else:
        return 0
train['is_weekend'] = train['dayofweek'].apply(isweekend)

def primetime(x):
    if(x>=6) and (x<=18):
        return 1
    else:
        return 0
train['prime_time'] = train['hour'].apply(primetime)

```

```python
train_new = train
def is_holiday(x):
    if x == None:
        return 0
    else:
        return 1
    
train_new['is_holiday'] = train['is_holiday'].apply(is_holiday)
```

```python
train_new1 = pd.get_dummies(train_new, columns = ['weather_description'])
```

```python
train_new1.columns
```

```python
test = train_new1['traffic_volume']

train_new2 = train_new1.drop(['date_time', 'weather_type', 'traffic_volume'], axis = 1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_new2, test, test_size = 0.33, random_state = 42)
```
